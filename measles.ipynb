```python
cols_of_interest = [
    'measles_suspect',
    'measles_clinical',
    'measles_epi_linked',
    'measles_lab_confirmed',
    'measles_total',
    'rubella_clinical',
    'rubella_epi_linked',
    'rubella_lab_confirmed',
    'rubella_total',
    'discarded'
]

regions = ['AFR', 'AMR', 'EMR', 'EUR', 'SEAR', 'WPR']

records = []

for region in regions:
    df_region = cases_month[cases_month['region'] == region]
    for col in cols_of_interest:
        idx = df_region[col].idxmax()
        row = df_region.loc[idx].copy()
        row['metric'] = col
        row['max_value'] = row[col]
        records.append(row)

region_max = pd.DataFrame(records)

region_max
```

```python
region_max['country'].unique()
```

```python
region_max['year'].min()
```

```python
most_common_month = (
    region_max.groupby('region')['month'].agg(lambda x: x.value_counts().idxmax()).reset_index(name='most_common_month')
)
most_common_month
```

```python
most_common_country = (
    region_max.groupby('region')['country'].agg(lambda x: x.value_counts().idxmax()).reset_index(name='most_common_country')
)
most_common_country
```
